# Manipulação de dados

Como existem diferentes tipos de interações, associamos a elas um peso ou força, assumindo que, por exemplo, que o tempo gasto numa página e a quantidade de vezes que a pessoa retornou a notícia indicam um interesse maior do usuário na notícia, do que um simples clique.

```python
interaction_weights = {
    'numberOfClicksHistory': 0.5,   # Clique indica interesse inicial
    'timeOnPageHistory': 1.0,       # Tempo na página indica engajamento mais forte
    'scrollPercentageHistory': 0.8, # Scroll ainda é relevante, mas menos que tempo na página
    'pageVisitsCountHistory': 1.2   # Visitas repetidas indicam alto interesse
}

def calculate_event_strength(row):
    """
    Calcula a pontuação de engajamento para cada interação, ponderando as métricas.
    """
    return (
        row['numberOfClicksHistory'] * interaction_weights['numberOfClicksHistory'] +
        row['timeOnPageHistory'] * interaction_weights['timeOnPageHistory'] +
        row['scrollPercentageHistory'] * interaction_weights['scrollPercentageHistory'] +
        row['pageVisitsCountHistory'] * interaction_weights['pageVisitsCountHistory']
    )

# Aplicando a função para criar a coluna 'eventStrength'
df_treino['eventStrength'] = df_treino.apply(calculate_event_strength, axis=1)
df_treino
```

### Problema de Cold Start do Usuário

Os sistemas de recomendação têm um problema conhecido como **user cold start**, no qual é difícil fornecer recomendações personalizadas para usuários com nenhum ou um número muito pequeno de itens consumidos, devido à falta de informações para modelar suas preferências.

Por esse motivo, estamos mantendo no conjunto de dados apenas usuários com pelo menos 5 interações (mais de 5 notícias lidas). Com poucas interações, não há dados suficientes para entender suas preferências. Isso evita recomendações aleatórias ou baseadas apenas em popularidade.

```python
df_treino = df_treino[df_treino['historySize'] >= 5]

# Verificar quantos usuários ficaram
print(f"Total de usuários após filtro: {df_treino['userId'].nunique()}")
```
